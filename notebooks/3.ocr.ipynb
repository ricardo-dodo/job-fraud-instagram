```python
import pandas as pd
import easyocr
import os
```

```python
# Membuat reader OCR menggunakan easyocr dengan bahasa Indonesia dan Inggris
reader = easyocr.Reader(['id', 'en'])

# Menentukan direktori tempat gambar disimpan
image_directory = 'downloaded_images/'

# Membaca file CSV
df = pd.read_csv('updated_ocr_unpreprocessed.csv')

# Menyiapkan kolom baru untuk hasil OCR
if 'OCR_Text' not in df.columns:
    df['OCR_Text'] = ''

for index, row in df.iterrows():
    # Memeriksa apakah hasil OCR sudah ada
    if pd.isna(row['OCR_Text']) or row['OCR_Text'] == '':
        image_path = os.path.join(image_directory, row['Filename'])
        if os.path.exists(image_path):
            print(f'Memproses gambar: {row["Filename"]}')  # Mencetak nama file yang diproses
            # Melakukan OCR pada gambar
            result = reader.readtext(image_path)
            # Menggabungkan teks yang dideteksi menjadi satu string dan menyimpan dalam DataFrame
            df.at[index, 'OCR_Text'] = ' '.join([text[1] for text in result])
            
            # Menyimpan DataFrame setelah setiap iterasi untuk memantau perkembangan
            df.to_csv('updated_ocr_unpreprocessed1.csv', index=False)
        else:
            print(f'Gambar tidak ditemukan: {row["Filename"]}')  # Jika gambar tidak ditemukan
    else:
        print(f'Sudah diproses: {row["Filename"]}')  # Jika hasil OCR sudah ada
```
