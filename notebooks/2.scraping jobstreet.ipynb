```python
import csv
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.common.exceptions import NoSuchElementException
import time
```

```python
# Setup WebDriver
service = ChromeService(executable_path=ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)
```

```python
driver.get("https://www.jobstreet.co.id/en/companies/browse-reviews/")
```

```python
companies = set()
```

```python
# Fungsi untuk mengambil data perusahaan dan rating
def scrape_data():
    company_elements = driver.find_elements(By.XPATH, "//a[@data-automation='CompanySearchResult']/span[@class='bmf23VXrbKovb5Hw_dat']")
    rating_elements = driver.find_elements(By.XPATH, "//div[@data-automation='CompanyDetails']//span[@class='bmf23VXrbKovb5Hw_dat']")

    for company, rating in zip(company_elements, rating_elements):
        company_info = (company.text, rating.text)  # Gunakan tuple untuk set
        if company_info not in companies:
            companies.add(company_info)
            print(company_info)  # Mencetak setiap perusahaan yang terambil
```

```python
def autoscroll_to_show_more_button():
    try:
        show_more_button = driver.find_element(By.XPATH, "//span[normalize-space()='Show more companies']")
        driver.execute_script("arguments[0].scrollIntoView();", show_more_button)
        return show_more_button
    except NoSuchElementException:
        return None
```

```python
# Panggil fungsi scrape_data untuk mengambil data pertama kali
scrape_data()
```

```python
# Cek dan klik tombol 'Show More Companies' jika ada, dengan autoscrolling
while (show_more_button := autoscroll_to_show_more_button()) is not None:
    show_more_button.click()
    autoscroll_to_show_more_button
    
   
```

```python
with open('company_ratings.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Company Name', 'Rating'])  # Header
    for company_info in companies:
        writer.writerow(company_info)
```
